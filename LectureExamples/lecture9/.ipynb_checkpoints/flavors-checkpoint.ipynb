```python
import datascience as ds
import numpy as np
%matplotlib inline
icecream = ds.Table().with_columns(
    'Flavor', ds.make_array('Chocolate', 'Strawberry', 'Vanilla'),
    'Number of Cartons', ds.make_array(16, 5, 9)
)
icecream
```

```python
icecream.barh('Flavor', 'Number of Cartons')
## Second label is optional if there are only 2 and the 2nd one is numeric
```

```python
icecream.sort('Number of Cartons', descending=True).barh('Flavor')  ## sort order
```

```python
top = ds.Table.read_table('data/' + 'top_movies.csv')
top
```

```python
movies_and_studios = top.select('Title', 'Studio')
movies_and_studios.group('Studio')
```

```python
studio_distribution = movies_and_studios.group('Studio')
studio_distribution.sort('count', descending=True).barh('Studio')
```

```python
movies_and_years = top.select('Title', 'Year')
movies_and_years.group('Year').sort('count', descending=True).barh('Year')
```
