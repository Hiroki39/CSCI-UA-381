```python
import datascience as ds
import numpy as np
%matplotlib inline
path = 'data/'
actors = ds.Table.read_table(path + 'actors.csv')
actors.show(actors.num_rows)
```

```python
actors.scatter('Number of Movies', 'Average per Movie')
```

```python
no_outlier = actors.where('Number of Movies', ds.are.above(10))
print('total lines',actors.num_rows,'above_10',no_outlier.num_rows)
print('Actors\n')

print('Removing outliers\n')
no_outlier.show(no_outlier.num_rows)
```

```python
no_outlier.scatter('Number of Movies', 'Average per Movie')
```

```python
above_60 = actors.where('Number of Movies', ds.are.above(60))
above_60.show(above_60.num_rows)
above_60.scatter('Number of Movies', 'Average per Movie')
```
